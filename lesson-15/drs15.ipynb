```python
d={
    "Region":["Tashkent","Tashkent","Tashkent","Tashkent","Samarkand","Samarkand","Samarkand"],
    "City":["Chilonzor","Yashnobod","Mirobod","Sergeli","Bulungur","Urgut","Jomboy"],
    "aholi":[5000,4000,8000,6000,2500,2000,1500]
}
df2=pd.DataFrame(d)
df2
```

```python
df3=df2.groupby("Region").agg({"aholi":["sum","min","max","mean"],"City":"count"})
```

```python
df2.groupby(["Region","City"]).agg({"aholi":"sum"})
```

```python
df2.groupby(["Region","City"]).agg({"aholi":"sum"}).reset_index()
```

```python
df3.columns
```

```python
for i in df3.columns:
    print("_".join(i))
```

```python
columns=[]
for i in df3.columns:
    columns.append("_".join(i))
columns
```

```python
df3.columns=columns
df3
```

```python
df3.reset_index(inplace=True)
df3
```

```python
df3.to_csv('countries.xlsx',index=False)
```

```python
d={
    "DepId":[1,2,3],
    "DepName":["Finanse","HR","Marketing"]
}
dep=pd.DataFrame(d)
dep
```

```python
emp.join(dep)
```

```python
emp.join(dep.add_suffix("_dept"))
```

```python
emp.set_index("DepId").join(dep.set_index("DepId"))    # join index ga qarab qo`shadi`
```
