```python
d={
    "product":["lap","lap","lap","phon","phon","phon","tab","tab","tab"],
    "category":["low","med","high","low","med","high","low","med","high"],
    "sales":[100,250,300,25,100,130,120,140,150]
}
df=pd.DataFrame(d)
df
```
